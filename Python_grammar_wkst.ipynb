## 五、 推导式

推导式是一种独特的数据处理方式，可以从一个数据序列构建另一个新的数据序列的结构体。

### 列表推导式 List Comprehension 
```
[out_exp_res for out_exp in input_list]
[out_exp_res for out_exp in input_list if condition]
```
实例：

````python
# 过滤掉长度小于或等于3的字符串列表，并将剩下的转换成大写字母
names = ['Bob','Tom','alice','Jerry','Wendy','Smith']
new_names = [name.upper()for name in names if len(name)>3]
print(new_names)
````

````python
# 计算 30 以内可以被 3 整除的整数
multiples = [i for i in range(30) if i % 3 == 0]
print(multiples)
````

### 字典推导式 Dictionary Comprehension
```
{ key_expr: value_expr for value in collection }
{ key_expr: value_expr for value in collection if condition }
```

````python
# 将列表转为字典（元素作为键，平方作为值）
numbers = [1, 2, 3, 4]
squared_dict = {x: x**2 for x in numbers}
print(squared_dict)
````

````python
# 将字符串列表转为字典（字符串作为键，长度作为值）
fruits = ["apple", "banana", "cherry"]
length_dict = {fruit: len(fruit) for fruit in fruits}
print(length_dict)
````

### 元组推导式
```
(expression for item in Sequence )
(expression for item in Sequence if conditional )
```
元组推导式和列表推导式的用法完全相同，只是元组推导式是用 `()` 圆括号将各部分括起来，而列表推导式用的是中括号 `[]` 。

### 集合推导式
```
{ expression for item in Sequence }
{ expression for item in Sequence if conditional }
```

````python
# 只保留偶数
numbers = [1, 2, 3, 3, 4, 5, 5, 6, 6, 6, 6]
even_numbers = {x for x in numbers if x % 2 == 0}
print(even_numbers)
````

````python
# 把字符变成对应的长度
words = ["apple", "banana", "cherry", "apple"]
lengths = {len(word) for word in words}
print(lengths)
````

## 六、迭代器与生成器
### 迭代器 Iterator (iter & next)
迭代器能让你逐个访问集合元素

它知道"下一个"是什么，但不知道"上一个"或"全部"

有点像这个东西（单向的链表）：

```
typedef struct Node {
    int data;
    struct Node *next;
} Node;
```

````python
my_list = [1, 2, 3]
it = iter(my_list)

print(next(it))  # 1
print(next(it))  # 2
print(next(it))  # 3
# 再next()就报错，因为没元素了
````

````python
# 用try避免一直next报错
my_list = [1, 2, 3]
it = iter(my_list)

while True:
    try:
        print(next(it))
    except StopIteration:
        print("end")
        break
````

好处：

1. 节省内存（不需要一次性加载所有数据）

2. 统一访问各种数据结构的接口（列表、字典、文件等都可用同样方式遍历）
### 生成器 Generator
#### 生成器函数（yield）

生成器本质上是一个 迭代器（Iterator），通过yield函数暂停迭代，并保存当前状态。

函数输出之后不会终止，而是暂停并保留所有局部状态，下次调用时从上次暂停的位置继续执行。（对比return）

````python
def count_to_infinity():
    n = 0
    while True:
        yield n
        n += 1

lazy_counter = count_to_infinity()  # 不会卡死你的电脑
print(next(lazy_counter))  # 0
print(next(lazy_counter))  # 1
# 可以一直next下去...
````

````python
# fibonacci, 避免绕脑子的recursion
def fibonacci(limit):
    a, b = 0, 1
    while a < limit:
        yield a
        a, b = b, a + b

for num in fibonacci(100):
    print(num)
````

````python
# 委托生成 yield from
def sub_generator():
    yield "Python"
    yield "Rust"

def main_generator():
    yield "Hello"
    yield from sub_generator()
    yield "World"

for word in main_generator():
    print(word)
````

#### 生成器表达式

````python
# 普通list
squares = [x*x for x in range(1000000)] # generate all at once，memory explode
````

````python
# 生成器表达式，只有在调用 next() 或迭代时，才会计算当前值
lazy_squares = (x*x for x in range(1000000))
print(next(lazy_squares))
print(next(lazy_squares))
````

### 类和对象

“类”就像“狗的品种”

比如“泰迪犬”是一个品种（类），它定义了这类狗的共同特点（名字、会叫）。

你家的“小白”是具体的某一只泰迪（对象），它是根据“泰迪犬”这个品种创造出来的。

想象你买了一只新狗狗，回家第一件事是给它起名字、戴项圈、放水碗。

`__init__` 就是“初始化”这只狗的过程——新狗到家时必须做的准备工作。

`__init__` 是构造函数，创建对象时自动调用。

self 代表这只狗自己（比如“小白的碗”就是 self.bowl）。

````python
class TeddyDog:
    def __init__(self, name):
        self.name = name
        self.bowl = "empty" # No dog food!   

    def bark(self):
        print(f"{self.name}: Wolf!")

dog1 = TeddyDog("Little_White")
dog2 = TeddyDog("Little_Black")
dog1.bark()
dog2.bark()
````

### Private and Public

Public：像狗狗的“名字”，谁都可以问：“你叫啥？” 

(属性/方法名无下划线（如 name），外部可直接访问。)

Private：像狗狗的“肠胃健康”，只有兽医（类内部）能检查，外人不能随便碰！

(属性/方法名加双下划线（如 __stomach），仅限内部使用，外部访问会报错)

````python
class TeddyDog:
    def __init__(self, name):
        self.name = name
        self.__stomach = "healthy"

    def check_health(self):
        print(f"{self.name}'s stomach: {self.__stomach}")

dog = TeddyDog("Little_White")
print(dog.name)
# print(dog.__stomach) # wrong!
dog.check_health()
````
