###  Dictionary Comprehension
```
{ key_expr: value_expr for value in collection }
{ key_expr: value_expr for value in collection if condition }
```

````python
# 
numbers = [1, 2, 3, 4]
squared_dict = {x: x**2 for x in numbers}
print(squared_dict)
````

````python
# 
fruits = ["apple", "banana", "cherry"]
length_dict = {fruit: len(fruit) for fruit in fruits}
print(length_dict)
````

###  Generator
#### yield

 Iteratoryield

return

````python
def count_to_infinity():
    n = 0
    while True:
        yield n
        n += 1

lazy_counter = count_to_infinity()  # 
print(next(lazy_counter))  # 0
print(next(lazy_counter))  # 1
# next...
````

````python
# fibonacci, recursion
def fibonacci(limit):
    a, b = 0, 1
    while a < limit:
        yield a
        a, b = b, a + b

for num in fibonacci(100):
    print(num)
````

````python
#  yield from
def sub_generator():
    yield "Python"
    yield "Rust"

def main_generator():
    yield "Hello"
    yield from sub_generator()
    yield "World"

for word in main_generator():
    print(word)
````
