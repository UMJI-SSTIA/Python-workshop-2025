```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 30, 10, endpoint=True)
y = np.array([3.242, 3.249, 3.249, 3.260, 3.270, 3.271, 3.273, 3.277, 3.267, 3.251])
y2 = np.array([0.275, 0.269, 0.273, 0.279, 0.275, 0.276, 0.270, 0.250, 0.262, 0.269])
```

```python
plt.plot(x, y, 'o-', label='Data Points')
plt.xlabel('t (s)')
plt.ylabel('U (V)')
plt.ylim(3.24, 3.305)
mean_y = np.mean(y)
plt.axhline(mean_y, color='red', linestyle='--', label=f'Mean = {mean_y:.3f} V')
std_y = np.std(y)
plt.fill_between(x, mean_y - std_y, mean_y + std_y, color='red', alpha=0.1, label='1 range')
plt.legend()
delta = np.max(y) - np.min(y)
plt.title(f'The Stability of the Voltage (U = {delta:.3f} V)')
plt.annotate(f'Max: {np.max(y):.3f} V', xy=(x[np.argmax(y)], np.max(y)), 
             xytext=(x[np.argmax(y)]+1, np.max(y)+0.01),
             arrowprops=dict(arrowstyle='->'))
for i in range(len(x)):
    plt.text(x[i], y[i] + 0.005, f'{y[i]:.3f}', ha='center', fontsize=8)
```

```python
plt.plot(x, y2, 'o-', label='Data Points')
plt.xlabel('t (s)')
plt.ylabel('I (A)')
plt.ylim(0.25, 0.30)
mean_y = np.mean(y2)
plt.axhline(mean_y, color='red', linestyle='--', label=f'Mean = {mean_y:.3f} A')
std_y = np.std(y2)
plt.fill_between(x, mean_y - std_y, mean_y + std_y, color='red', alpha=0.1, label='1 range')
plt.legend()
delta = np.max(y2) - np.min(y2)
plt.title(f'The Stability of the Current (I = {delta:.3f} A)')
plt.annotate(f'Max: {np.max(y2):.3f} A', xy=(x[np.argmax(y2)], np.max(y2)), 
             xytext=(x[np.argmax(y2)]+1, np.max(y2)+0.01),
             arrowprops=dict(arrowstyle='->'))
for i in range(len(x)):     
    plt.text(x[i], y2[i] + 0.005, f'{y2[i]:.3f}', ha='center', fontsize=8)                  
```
